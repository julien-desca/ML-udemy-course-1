```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv('paris_airbnb.csv')
```

```python
dataframe
```

# Calculer la distance euclidienne
- Calculer la distance entre chaque valeur de la colonne 'acommodates' du dataframe et la valeur 3 (nom de personne que peut acceuillir notre logement)
- Assigner les valeurs de distance à la conne 'distance' du dataframe
- Utiliser la méthode Series value_counts() et afficher le décompte de chaque valeur unique de la colonne 'distance'

```python
distances = dataframe['accommodates'].apply(lambda x: np.abs(x-3))
dataframe['distance'] = distances
dataframe['distance'].value_counts()
```

# Randomiser le dataset
- Placer le seed : np.random.seed(1)
- Randomiser l'ordre des lignes du dataframe
    - np.random.permutation() pour retourner un tableau Numpy de valeurs d'index mélangés
    - dataframe.loc[] pour retourner un nouveau dataframe dans l'ordre mélangé
- trier dataframe sur la colonne distance
- afficher les 10 premières valeur de la colonne 'price

```python
np.random.seed(1)
indexes = np.random.permutation(len(dataframe))
dataframe = dataframe.loc[indexes]
dataframe = dataframe.sort_values('distance')
dataframe['price']
```

```python
dataframe['formatted_price'] = dataframe['price'].apply(lambda x : x.replace('$', '').replace(',', '')).astype(float)
dataframe.head()['formatted_price'].mean()
```
